```python
def has_return(arg):
    if arg == None:
        return("N/A")
    else:
        return arg.text
```

```python
#solution 3
#จัดการปัญหา ไม่มีค่าในบางฟิลด์ 
#ยังพบปัญหา ที่ระดับ page level เนื่องจากเว็บไซต์ fetch ไม่ทัน ในบางหน้า ด้วย try catch
hotel_list = []
total_error = 0
for i in range(0,len(pages),1):
    print(i)
    soup = bs4.BeautifulSoup(pages[i])
    try:
        hotel_container = soup.find_all('li',{'class':'PropertyCard PropertyCardItem','data-selenium':'hotel-item'})
        for j in range(0,len(hotel_container),1):
            hotel_elements = hotel_container[j].find('div',{'class':'Gridstyled__GridStyled-sc-dfh2k0-0 kEJDDi PropertyCard__Container'})
        #ส่วนดึงข้อมูล
            try:
                hotel_name = hotel_elements.find('h3',{'class':'sc-jrAGrp sc-kEjbxe eDlaBj dscgss','data-selenium':'hotel-name'})               
                room_type = hotel_elements.find('span',{'class':'sc-jrAGrp sc-kEjbxe eDlaBj jwmMlM'})
                review_text = hotel_elements.find('span',{'class':'Spanstyled__SpanStyled-sc-16tp9kb-0 kkSkZk kite-js-Span Box-sc-kv6pi1-0 eRxXoo'})
                review_score = hotel_elements.find('p',{'class':'Typographystyled__TypographyStyled-sc-j18mtu-0 Hkrzy kite-js-Typography'})
                review_users = hotel_elements.find('span',{'class':'Spanstyled__SpanStyled-sc-16tp9kb-0 jYmZbG kite-js-Span Box-sc-kv6pi1-0 jjmSNA'})
                price = hotel_elements.find('span',{'class':'PropertyCardPrice__Value'})

                hotel_name = has_return(hotel_name)
                room_type = has_return(room_type)
                review_text = has_return(review_text)
                review_score = has_return(review_score)
                review_users = has_return(review_users)
                price = has_return(price)
                print(j,' ',hotel_name)
    
                hotel_list.append([hotel_name,room_type,review_score,review_text,review_users,price])
            except Exception as error:
                print('Error at subpage level #',i,' #',j)
                print("An error occurred:", error) 
                total_error = total_error + 1

    except Exception as error:
        print('Error at page level #',i,' #',j)
        print("An error occurred:", error) 
#         total_error = total_error + 50 #เนื่องจาก ในแต่ละหน้ามี 50 hotels

print('found error = ',total_error)        
```
