```python
#‡∏î‡∏∂‡∏á html source ‡πÄ‡∏Å‡πá‡∏ö‡πÑ‡∏ß‡πâ
data = driver.page_source
```

```python
soup = bs4.BeautifulSoup(data)
```

```python
#‡∏Å‡∏≤‡∏£‡∏ó‡∏≥‡∏á‡∏≤‡∏ô‡∏Å‡∏≤‡∏£ soup ‡πÉ‡∏ä‡πâ‡∏Å‡∏≤‡∏£‡∏£‡∏∞‡∏ö‡∏∏ tag / class ‡πÅ‡∏ó‡∏ô ‡∏Å‡∏≤‡∏£ copy xpath

products = soup.find_all('div',{'class':'ie3A+n bM+7UW Cve6sh'}) #product name
prices = soup.find_all('div',{'class':'vioxXd rVLWG6'}) #price
amounts = soup.find_all('div',{'class':'r6HknA uEPGHT'}) #amount

#!!!!!!!!!!!!!!!!!!‡∏¢‡∏±‡∏á‡∏°‡∏µ‡∏õ‡∏±‡∏ç‡∏´‡∏≤‡πÄ‡∏£‡∏∑‡πà‡∏≠‡∏á ‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤‡∏ó‡∏µ‡πà‡∏Ç‡∏≤‡∏¢‡πÑ‡∏°‡πà‡πÑ‡∏î‡πâ ‡∏à‡∏∞‡πÑ‡∏°‡πà‡∏ñ‡∏π‡∏Å‡∏Ñ‡∏±‡∏î‡∏•‡∏≠‡∏Å‡∏°‡∏≤ ‡∏î‡∏±‡∏á‡∏ô‡∏±‡πâ‡∏ô ‡∏™‡∏¥‡∏ô‡∏Ñ‡πâ‡∏≤ ‡∏à‡∏∂‡∏á‡∏°‡∏µ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏ó‡∏µ‡πà‡∏Ç‡∏≤‡∏¢‡πÑ‡∏°‡πà‡∏Ñ‡∏£‡∏ö
```

```python
#convert to text
for product in products:
    print(product.text)
```

```python
#convert to list
product_list = []

for product in products:
    product_list.append(product.text)
product_list
```

```python
prices_list = []
for price in prices:
    prices_list.append(price.text)
prices_list
```

```python
amounts_list = []
for amount in amounts:
    amounts_list.append(amount.text)
amounts_list
```

```python
import pandas as pd
```

```python
shopee_data = pd.DataFrame([product_list,prices_list,amounts_list])
```

```python
shopee_data = shopee_data.transpose()
```

```python
shopee_data.columns = ['title','price','amount']
```
